```python
#W4A6
try:
    a = int(input("nhap a "))
    b = int(input("nhap b "))

    if a < 1 or b < 1 or a > b:
        print("nhap lai")
    else:
        sum = 0
        for num in range(a, b + 1):
            if num > 1:
                flag = True
                for i in range(2, num):
                    if num % i == 0:
                        flag = False
                        break

                if flag:
                    sum += num

        print(sum)

except ValueError:
    print("nhap lai")
```

```python
#W4A8
try:
    n = int(input("nhap n"))
except ValueError:
    print("nhap lai")
    exit()
if n <= 0:
    print("nhap lai")
    exit()
buoc = 0
so_hien_tai = n
while True:
    s_hien_tai = str(so_hien_tai)
    s_dao_nguoc = s_hien_tai[::-1]
    if s_hien_tai == s_dao_nguoc:
        break
    so_dao_nguoc = int(s_dao_nguoc)
    so_hien_tai = so_hien_tai + so_dao_nguoc
    buoc += 1
print(f"so buoc {buoc}")
print(f"so doi xung la {so_hien_tai}")
```

```python
#W4A9
try:
    n = int(input("nhap n nguyen duong "))

    if n <= 0:
        print("nhap lai")
    else:
        lst= []
        k = 1

        while k * k <= n:
            a = k * k
            at = str(a)
            flag= True
            for i in range(len(at)):
                for j in range(i + 1, len(at)):
                    if at[i] == at[j]:
                        flag = False
                        break

                if not flag:
                    break

            if flag:
                lst.append(a)

            k += 1

        if lst:
            print(*lst)
        else:
            print("(no number)")

except ValueError:
    print("nhap lai")
```

```python
#W4A10
try:
    n = int(input("nhap n nguyen duong "))
except ValueError:
    print("nhap lai")
    exit()
if n <= 0:
    print("nhap lai")
    exit()
ddln = 0
best_x = 0
for x in range(1, n + 1):
    sht = x
    ddht = 1
    while sht != 1:
        if sht % 2 == 0:
            sht = sht // 2
        else:
            sht = 3 *sht + 1
        ddht+= 1
    if ddht > ddln:
        ddln = ddht
        best_x = x
print(f"{best_x} {ddln}")
```

```python
#W4A13
a=int(input())
b=int(input())
lsta=[]
lstb=[]
suma=0
sumb=0
for i in range(1,a+1):
  if a%i==0:
    lsta.append(i)
for v in range(1,b+1):
  if b%v==0:
    lstb.append(v)
for x in lsta:
  suma+=x
for y in lstb:
  sumb+=y
if suma-a==b and sumb-b==a:
  print("yes")
else:
  print("no")
```
