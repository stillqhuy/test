```python
# w9a1 Viết hàm merge hai danh sách đã sắp xếp mà không dùng bất cứ thuật toán sắp xếp nào
l1=list(map(int,input().split()))
l2=list(map(int,input().split()))
i=0
j=0
kq=[]
while i<len(l1) and j<len(l2):
    if l1[i]>l2[j]:
        kq.append(l2[j])
        j+=1
    else:
        kq.append(l1[i])
        i+=1
kq+=l1[i:]
kq+=l2[j:]
print(kq)
    

```

```python
# w9a2Cho ma trận hai chiều nxm có các phần tử được sắp xếp trên dòng tăng dần và trên cột tăng dần, hãy tìm số lớn thứ k (k < 
# nxm) với thuật toán tốt nhất có thể 
matrix=[]
hang=int(input("nhap so hang"))
cot=int(input("nhap so cot"))
k=int(input("nhap k"))
for i in range(hang):
    l=[]
    for j in range(cot):
        l.append(int(input()))
    matrix.append(l)
print(matrix)
phang=[]
for hang in matrix:
    for i in hang:
        phang.append(i)
phang.sort(reverse=True)
print(phang[k-1])
        

```

```python
# w9a5 Cho hai mảng số nguyên list1 và list2, viết hàm kiểm tra xem list1 
# có thuộc list2 hay không (có phải là dãy con liên tiếp của list2 hay 
# không)?
l1=list(map(int,input().split()))
l2=list(map(int,input().split()))
flag=False
if len(l1)>len(l2):
    flag=False
for i in range (len(l2)):
    if l1==l2[i:i+len(l1)]:
        flag=True
        break
if flag:
    print("YES")
else:
    print("NO")
```

```python
vitri=list(map(int,input().split()))
socay=int(input())
pos=0
for i in range(1,len(vitri)-1):
    if abs(vitri[i]-vitri[i+1])!=1 and abs(vitri[i]-vitri[i-1])!=1:
        pos+=1
if pos>=socay:
    print("yes")
else:
    print("no")
```
