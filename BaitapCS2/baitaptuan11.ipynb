```python
# w11a1
# Đề bài: Mọi số nguyên dương đều có một số đảo ngược. Hãy đọc một số từ file và trả về số đảo ngược của nó.
# Yêu cầu hàm: def reverse(path: str) -> int
# Input (File): Chứa duy nhất một số nguyên dương.
# Output: Số nguyên đã đảo ngược.
def reverse(l):
    with open(l,"r") as fi:
        data=fi.read().strip()
        return int(data[::-1])
with open("f","w") as fu:
    fu.write("123456")
print(reverse("f"))
    
```

```python
# w11a4 
# Đề bài: Kiểm tra xem nội dung của 2 file có phải là đảo chữ (anagram) của nhau không? (Anagram là dùng các chữ cái của từ này sắp xếp lại thành
# từ kia). Không phân biệt hoa thường, bỏ qua dấu câu.
# Yêu cầu hàm: def areAnagrams(path1: str, path2: str) -> bool
# Ví dụ:
# File 1: Tom Marvolo Riddle
# File 2: I Am Lord Voldemort
# Output: True
from pathlib import Path
WORKDIR=Path(".")
path1="str1"
path2="str2"
with open(path1,"w") as f:
    f.write("bui quang huy")
with open (path2,"w") as a:
    a.write("hyu q uangbui")
with open (path1,"r") as p1:
    ch1=p1.read()
with open (path2,"r") as p2:
    ch2=p2.read()
l1=[]
l2=[]
for i in ch1:
    if i.isalnum():
        l1.append(i)
for i in ch2:
    if i.isalnum():
        l2.append(i)
l1.sort()
l2.sort()

if l1==l2:
    print(True)
else:
    print(False)

```

```python
#w11a6
path_file = 'w11a6.txt'
noi_dung_mau = "Everyone deserves a second shot."
with open(path_file, "w") as f:
    f.write(noi_dung_mau)
def countCharacter(path, letter):
    count = 0
    target = letter.lower() 
    with open(path, "r") as f:
        content = f.read() 
    for char in content:
        if char.lower() == target:
            count += 1
    return count
ky_tu_can_tim = 'e' 
ket_qua = countCharacter(path_file, ky_tu_can_tim)
print(f"Văn bản: '{noi_dung_mau}'")
print(f"Số lần xuất hiện của kí tự '{ky_tu_can_tim}': {ket_qua}")
```

```python
# w11a15
# Thành phố X mở một rạp chiếu phim mới, rất nhiều người muốn đến xem thử và hưởng các khuyến mại trong ngày khai trương.
# Để đưa cho khách hàng các lựa chọn, rạp chiếu phim treo các poster về phim cùng với điểm đánh giá IMDB (rating) và mô tả của khách hàng về bộ phim đó.
# Hãy viết hàm def findMovies (path: str) -> list[str] giúp khán giả tìm ra phim với điểm đánh giá lớn hơn 8.0 và có mô tả Không phải là "nhàm chán" ("boring")
# Hàm nhận đầu vào là đường dẫn đến tệp chứa các poster phim. Hàm trả lại các tên những phim thỏa mãn điều kiện trên. Nếu không tìm thấy phim nào phù hợp, trả 
# về thông báo "All are boring!". Kết quả in theo thứ tự xuất hiện trước sau trong file đầu vào
# Tệp có cấu trúc như sau:
# id, movie, description, rating
# 1, Thor, great 3D, 8.0
# 2, The Rings, Fantasy, 7.5
# 3, Irish, boring, 6.2
# 4, Ice song, Fantasy, 8.6
# 5, Home Alone, Interesting, 6.5
import csv
def findMovies(path):
    kq = []
    with open(path, 'r') as f:
        reader =csv.reader(f, skipinitialspace=True)
        next(reader) 
        for row in reader:
            if float(row[3]) > 8.0 and row[2] != 'boring':
                kq.append(row[1])
    if len(kq) == 0:
        return ["All are boring!"]
    return kq
print(findMovies("w11a16.csv"))

```

```python
# w11a16
import math
def calc(p):
 with open(p)as f:
  for b in f.read().strip().split('\n\n'):
   l=b.split('\n')
   t=l[0].split()[1]
   v=[float(x)for x in l[1].replace('SIDE:','').replace('WIDTH:','').replace('HEIGHT:','').replace('RADIUS:','').split()]
   if t=='SQUARE':
    print(f"{v[0]*4:.2f}")
   elif t=='RECTANGLE':
    print(f"{(v[0]+v[1])*2:.2f}")
   elif t=='CIRCLE':
    print(f"{2*math.pi*v[0]:.2f}")
calc("w11a16")
```
