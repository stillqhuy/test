```python
# w5a8
# Khoảng cách Hamming giữa hai số nguyên là số lượng vị trí khác nhau giữa hai dãy bits 
# tương ứng của chúng. Ví dụ: Khoảng cách Hamming giữa 1 (1) và 4 (100) là 2. 
# Cho hai số nguyên x và y, hãy viết hàm trả về khoảng cách Hamming giữa hai số này.
x,y=map(int,input().split())
def nhiphan(a):
    l=[]
    if a==0:
        return [0]
    while a>0:
        l.append(a%2)
        a=a//2
    
    return l[::-1]
c=nhiphan(x)
v=nhiphan(y)
for i in range(max(len(c),len(v))-min(len(c),len(v))):
    if len(c)==min(len(c),len(v)):
        c.insert(0,i)
    else:
        v.insert(0,i)


def diff(c,v):
    h=0
    for i in range(len(c)):
        if c[i]==v[i]:
            h+=1
    return max(len(c),len(v))-h
print(diff(c,v))


```

```python
c,v=map(int,input().split())
def hamming(c,v):
    d=c^v
    sum=0
    while d > 0:
        if d % 2 == 1: 
            sum+= 1
        d = d // 2
    return sum
print(hamming(c,v))

```

```python
# w5A9 Viết hàm trả về tổng các chữ số của số nguyên dương n 
n=int(input())
def tong(n):
    l=[]
    while n>0:
        l.append(n%10)
        n=n//10
    return sum(l)
print(tong(n))

```
