```python
l=[1,2,3]
n=[1,3,4,5,6]
c=0
maxlen=max(len(l),len(n))
minlen=min(len(l),len(n))
c+=maxlen-minlen
for i,j in zip(l,n):
    if i!=j:
        c+=1
print(c)
```

```python
# w6a4
l=list(map(str,input().split()))
v=[]
d={}
for i in l:
    v.append(i.split(":"))
for i in v:
    key=i[0]
    value=i[1]
    if key not in d:
        d[key]=[value]
    else:
        d[key].append(value)
print(d)
        

```

```python
# w6a5 nhap vao mot list so nguyen in ra dic{'duong':,'am':,'khong':}
l=list(map(int,input().split()))
a={"duong":[],"am":[],"khong":[]}
lduong=[]
lam=[]
lkhong=[]
a['duong']=lduong
a['am']=lam
a['khong']=lkhong
for u in l:
    if u>0:
        lduong.append(u)
    elif u<0:
        lam.append(u)
    else:
        lkhong.append(u)
print(a)
```

```python
# w6a8
l=list(map(str,input().split()))
d={}
count=0
k=[]
for i in l:
    if i not in k:
        k.append(i)
print(k)
for i in k:
    for j in l:
        if i==j:
            count+=1
            d[i]=count
    count=0

print(d)
```

```python
#w6a9
d1=list(input().split())
d2=list(input().split())
caccap1=[]
caccap2=[]
for i,j in zip(d1,d2):
    cap1=i.split(":")
    caccap1.append(cap1)
    cap2=j.split(":")
    caccap2.append(cap2)

dic1=dict(caccap1)
dic2=dict(caccap2)
for k,v in dic2.items():
    if k in dic1:
        dic1[k]=[dic1[k],v]
    else:
        dic1[k]=v
print(dic1)

```

```python
# w6a10
l=list(map(int,input().split()))
k=int(input())
caccap=[]      
for i in range(len(l)):
    for j in range(i+1,len(l)):
        if l[i]+l[j]==k:
            caccap.append((l[i],l[j]))
print(caccap)
            
        
        
```

```python
# w6a15
l=list(map(str,input().split()))
k=int(input())
v=[]
for i in l:
    v.append(i.split(":"))
d=dict(v)
kq={}
print(d)
for i,j in zip(d.values(),d.keys()):
    if int(i)>k:
        kq[j]=int(i)
print(kq)
```
