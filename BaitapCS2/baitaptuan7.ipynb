```python
# W7A1: Tìm kiếm nhị phân
# Đề bài: Viết hàm tìm kiếm nhị phân trên một danh sách đã được sắp xếp tăng dần. 
# Trả về chỉ số của phần tử hoặc -1 nếu không tìm thấy. Mẫu hàm: def binary_search(arr, left, right, target):
arr=[1,2,3,4,5,6,7]
target=5
def binary_search(arr,target):
    left,right=0,len(arr)-1
    while left<=right:
        mid=(left+right)//2
        if arr[mid]==target:
            return mid
        elif arr[mid]<target:
            left=mid+1
        else:
            right=mid-1
    return -1
print(binary_search(arr,target))
```

```python

# W7A2 Đề bài:Cho danh sách arr và một số nguyên x, đếm xem x xuất hiện bao nhiêu lần trong danh sách.
# Mẫu hàm: def count_occurrences(arr, x) -> int:
arr=[1,4,1,3]
x=1
def countoc(arr,x):
    c=0
    for i in arr:
        if i==x:
            c+=1
    return c
print(countoc(arr,x))



```

```python
# W7A3Viết hàm thực hiện thuật toán bubble sort, Selection Sort để sắp xếp danh sách theo thứ tự tăng dần
arr=[9,7,3,5,1]
def bubblesort(arr):
    for j in range(len(arr)):
        for i in range(len(arr)-1):
            if arr[i]>arr[i+1]:
                arr[i],arr[i+1]=arr[i+1],arr[i]
    return arr
print(bubblesort(arr))
```

```python
# W7A5
# Cho danh sách các số nguyên và một số X, đếm tất cả các cặp (a, b) sao cho a + b = X
l=list(map(int,input().split()))
t=int(input())
def countc(l,t):
    c=0
    for i in range(len(l)):
        for j in range(i+1,len(l)):
            if l[i]+l[j]==t:
                c+=1
    return c
print(countc(l,t))
```

```python
# W7A7 Cho một danh sách các số nguyên nums, hãy tìm độ dài của dãy con tăng nghiêm ngặt dài nhất (không nhất thiết phải liền kề). 
l=list(map(int,input().split()))
print(l)
dp=[1]*len(l)
for i in range(1,len(l)):
    for j in range(i):
        if l[i]>l[j]:
           dp[i]=max(dp[i],dp[j]+1)
print(max(dp))
```

```python
# W7A8 Bạn được cung cấp: một danh sách các khoảng thời gian interval = [[l1, r1], [l2, r2], ...] và một danh sách các truy vấn queries = [q1, q2, ...]  
# Với mỗi truy vấn q, hãy tìm độ dài nhỏ nhất của khoảng thời gian [li, ri] sao cho li ≤ q ≤ ri.  Nếu không tồn tại khoảng thời gian nào như vậy, hãy 
# trả về -1 cho truy vấn đó. solve in < O(m.n), hint: binary search in the queries list
n=int(input("nhap so khoang thoi gian"))
interval=[]
for i in range(n):
    interval.append(list(map(int,input().split())))
queries=list(map(int,input("nhap cac truy van").split()))
kq=[]
def dodai(j):
    return j[-1]-j[0]+1
for i in queries:
    checkmin=[]
    for j in interval:
        if i in range (j[0],j[-1]+1):
            checkmin.append(dodai(j))
    if checkmin:
        kq.append(min(checkmin))
    else:
        kq.append(-1)
            
print(kq)


    

```

```python
# W7A9 Cho một danh sách các số nums, với mỗi phần tử nums[i], hãy đếm xem có bao nhiêu phần tử bên phải nhỏ hơn nums[i].  Trả về danh sách các số 
# đếm. Efficient solution: using heap or binary search tree
l=list(map(int,input().split()))
kq=[]
for i in range(len(l)):
    c=0
    for j in range(i+1,len(l)):
        if l[i]>l[j]:
            c+=1
    kq.append(c)
print(kq)
```

```python
# w7a10 Cho một danh sách các số nguyên nums và một số nguyên m, hãy tìm tổng mảng con lớn nhất theo modulo m.  Tức là, với bất kỳ mảng con nums[i:j] nào, hãy tính 
# sum(nums[i:j]) % m và trả về kết quả lớn nhất có thể. Gợi ý: Duy trì tổng tiền tố (prefix sum) mod m và sử dụng tìm kiếm nhị phân để tìm tổng tiền tố nhỏ nhất chỉ 
# lớn hơn tổng hiện tại một chút (để tối đa hóa hiệu số bao quanh).
l=list(map(int,input().split()))
n=int(input())
s=[]
for i in range(len(l)):
    for j in range(i+1,len(l)):
        s.append(sum(l[i:j]))
max=0
for i in range(len(s)):
    if s[i]%n>max:
        max=s[i]%n
for i in range (len(s)):
    if s[i]%n==max:
        print(s[i])
        break
```
