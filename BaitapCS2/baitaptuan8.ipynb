```python
# w7a17Tìm phần tử xuất hiện nhiều hơn n/2 lần trong mảng (nếu có) với độ phức tạp thời gian là O(n), độ phức tạp bộ nhớ là O(1), - Boyer–Moore. Nếu không có phần tử 
# nào thỏa mãn thì không in gì 
n=int(input())
l=list(map(int,input().split()))
ds=[0]*int(1e6)
for i in l:
    ds[i]+=1
for i in range(len(ds)):
    if ds[i]>n/2:
        print(i)
```

```python
# w7a18 Sắp xếp có điều kiện: Số chẵn đứng trước, số lẻ đứng sau, chẵn tăng dần, lẻ giảm dần
l=list(map(int,input().split()))
chan=[]
le=[]
for i in l:
    if i%2==0:
        chan.append(i)
    else:
        le.append(i)
chan.sort()
le.sort(reverse=True)
print(chan+le)
    
```

```python
# w7a19 Merge 2 mảng đã sắp xếp (có thể tăng dần, có thể giảm dần) mà không dùng bất cứ thuật toán sắp xếp nào thành một 
# mảng cũng có cùng thứ tự sắp xếp như 2 mảng ban đầu
l1=list(map(int,input().split()))
l2=list(map(int,input().split()))
i=0
j=0
l=[]
while i<len(l1) and j<len(l2):
    if l1[i]<=l2[j]:
        l.append(l1[i])
        i+=1
    else:
        l.append(l2[j])
        j+=1
l.extend(l1[i:])
l.extend(l2[j:])
print(l)

    
    
```

```python
# w7a20 Cho dãy N số nguyên dương, tính tổng các số có k chữ số. Dòng đầu ghi N số nguyên dương, dòng sau ghi số nguyên dương k
l=list(map(int,input().split()))
k=int(input())
p=[]
for i in l:
    if 1<=i/(10**(k-1))<10:
        p.append(i)
print(p)
print(sum(p))
```

```python
# w7a22 Cho N tên sinh viên gồm cả họ và tên, tìm xem tên nào hay gặp nhất. Nếu có nhiều tên cùng số lần gặp thì lấy tên có hứ tự từ điển 
N=int(input())
l=[]
for i in range(N):
    l.append(input())
for i in range(N):
    l[i]=l[i].split()
    l[i]=l[i][-1]
l.sort()
ds={}
for i in l:
    if i in ds:
        ds[i]+=1
    else:
        ds[i]=1
max_val=max(ds.values())
for i,j in ds.items():
    if j==max_val:
        print(i)




    
```

```python
# w7a25Tìm số có đúng k chữ số lớn nhất có thể trong một số có n chữ số (coi đầu vào là một số rất lớn - str). Không được hoán 
# đổi vị trí các chữ số
n=input()
l=list(n)
k=int(input())
cx=len(l)-k
st=[]
for c in l:
    while len(st)!=0 and st[-1]<c and cx>0:
        st.pop()
        cx-=1
    st.append(c)
if cx>0:
    for i in range(cx):
        st.pop()
for i in st:
    print(i,end='')
```

```python
# w7a28 Bà ̀ i toán Tìm dãy các hợp số dài nhất liên tiếp không chứa số nguyên nào <= N (N là số nguyên dương cho trước). Nếu có 
# nhiều dãy cùng độ dài thì lấy dãy nhỏ nhất (gặp đầu tiên từ trái qua) 
n=int(input())
from math import sqrt as s
def snt(a):
    for i in range(2,int(s(a)+1)):
        if a%i==0:
            return False
    return True
hs=[]
for i in range(1,n+1):
    if not snt(i) and i>1:
        hs.append(i)
print(hs)
temp=[]
kq=[]
temp.append(hs[0])
for i in range(1,len(hs)):
    if hs[i]==hs[i-1]+1:
        temp.append(hs[i])
    else:
        kq.append(temp)
        temp=[hs[i]]
kq.append(temp)
print(kq)
maxc=0
for i in kq:
    if len(i)>maxc:
        maxc=len(i)
for i in kq:
    if len(i)==maxc:
        print(i)
        break


    
    




```
