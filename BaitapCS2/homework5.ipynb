```python
# w6a28
l=list(map(int,input().split()))
def kiemtra(l):
    t=sorted(l)
    n=t[::-1]
    if len(set(l))!=len(l):
        return "vo danh"
    if l==t:
        return "day tang"
    elif l==n:
        return "day giam"
    else:
        return "vo danh"
print(kiemtra(l))

    
```
